```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
cf_tweets = pd.read_csv("Data/aggregate_results_24_07_2018_cleaned.csv")
```

```python
cf_tweets.columns
```

```python
cf_tweets = cf_tweets[['id', 'created_at', 'text', 'protest', 'violence']]
```

```python
cf_tweets["created_at_dt"] = pd.to_datetime(cf_tweets["created_at"])
```

```python
cf_tweets_dt = cf_tweets.set_index(cf_tweets["created_at_dt"])
```

```python
cf_tweets_dt["created_at_dt"].describe()
```

```python
len(cf_tweets_dt.resample("D"))
```

```python
len(cf_tweets_dt.resample("H"))
```

```python
plt.figure(figsize=(20,4))

plt.plot(cf_tweets_dt['violence'].resample("D", how='count'))
plt.show()
```

```python
plt.figure(figsize=(20,4))

plt.plot(cf_tweets_dt[cf_tweets_dt['violence']==1].resample("D", how='count'))
plt.show()
```

```python
plt.figure(figsize=(15,4))

plt.plot(cf_tweets_dt[cf_tweets_dt['violence']==1].resample("H", how='count'))
plt.show()
```
